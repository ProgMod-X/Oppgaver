### Oppgave 1.3.1.1
Bruk en while-løkke til å skrive ut tallene fra 1 til 20 i stigende rekkefølge.

### Oppgave 1.3.1.3
Beskriv forskjellen på en for-løkke og en while-løkke. Vis et eksempel på hvordan du kan skrive den ene typen om til den andre.

```python
# En for-løkke blir vanlig vis brukt til å iterere over elementene i en liste.
# En while-løkke kjører så lenge betingelsen er sann.

liste = list(range(5)) # dette lager listen [0, 1, 2, 3, 4]

# Nedenfor blir variabelen "i" brukt til å referere til elementet som loopen har kommet til.
for i in liste:
    print(i)
```

```python
# Ofte bruker man range() funksjonen for å raskt lage sekvens med en spesifikk lengde som loopen kan iterere over.
for i in range(3):
    print(i)
```

```python
# Hvis man skriver en for-loop inni en annen for-loop er det vanlig å bruke bokstavene j, k, l.
for i in range(5):
    for j in range(3):
        print(i, j)
    print()
```

```python
# While løkken kjører så lenge betingelsen er sann.
while input("Skal loopen fortsette:") == "ja":
    print("loop")
```

```python
# Ofte har man en variabel som teller antall repetisjoner, den blir ofte brukt til å bestemme hvor mange ganger loopen skal kjøre.
repetisjoner = 0
while repetisjoner < 5:

    print("Repetisjoner:", repetisjoner)

    repetisjoner += 1
```

```python
# Vi kan skrive en While-løkke som også itererer over elementene i en liste.

liste = ("eple", "banan", "tomat")

# Vi vil at listen skal kjøre like mange ganger som det er elementer i listen.
max = len(liste)

# Som vanlig bruker vi variabelen i til å telle hvor vi er i listen, vi starter på index 0.
i = 0

while i < max:
    element = liste[i]
    print(element)
    i += 1
```

### Oppgave 1.3.1.4
a) Bruk for-løkker til å skrive ut

    1) tallene 1 til 15 horisontalt, med et mellomrom mellom tallene

    2) tallene 1 til 15 vertikalt

    3) oddetallene 1 til 15 horisontalt, med et mellomrom mellom tallene

b) Skriv om løkkene over til  while-løkker.

```python
tall = ""
# lager en string med alle tallene
for i in range(1, 16):
    tall += " " + str(i)

print(tall)

for i in range(1, 16):
    print(i)

print()

for i in range(1, 16, 2):
    print(i)
```

```python
tall = ""
# lager en string med alle tallene
i = 1
while i <= 15:
    tall += " " + str(i)
    i += 1

print(tall)

i = 1
while i <= 15:
    print(i)
    i += 1

print()

i = 1
while i <= 15:
    print(i)
    i += 2
```

### Oppgave 1.3.1.5
Pyramidetall n er summen av alle kvadrattall fra og med 1 til og med $ n^2 $. Pyramidetall 4 er  visualisert under.



(Bilde: Wikipedia)

Skriv et program som spør brukeren om hvilket tall hen ønsker, og bruker én løkke til å beregne tallet. Skriv ut tallet

```python
n = int(input("n: "))
sum = 0

i = 1
while i <= n:
    sum += i ** 2
    print("lag: " + str(i), "\tlag størrelse: " + str(i**2) + "\tsum: " + str(sum))
    i += 1

print("\nPyramide tall n =", sum)
```
