### Oppgave 1.3.1.3
Beskriv forskjellen på en for-løkke og en while-løkke. Vis et eksempel på hvordan du kan skrive den ene typen om til den andre.

```python
# En for-løkke blir vanlig vis brukt til å iterere over elementene i en liste.
# En while-løkke kjører så lenge betingelsen er sann.

liste = list(range(5)) # dette lager listen [0, 1, 2, 3, 4]

# Nedenfor blir variabelen "i" brukt til å referere til elementet som loopen har kommet til.
for i in liste:
    print(i)
```

```python
# Ofte bruker man range() funksjonen for å raskt lage sekvens med en spesifikk lengde som loopen kan iterere over.
for i in range(3):
    print(i)
```

```python
# Hvis man skriver en for-loop inni en annen for-loop er det vanlig å bruke bokstavene j, k, l.
for i in range(5):
    for j in range(3):
        print(i, j)
    print()
```

```python
# While løkken kjører så lenge betingelsen er sann.
while input("Skal loopen fortsette:") == "ja":
    print("loop")
```

```python
# Ofte har man en variabel som teller antall repetisjoner, den blir ofte brukt til å bestemme hvor mange ganger loopen skal kjøre.
repetisjoner = 0
while repetisjoner < 5:

    print("Repetisjoner:", repetisjoner)

    repetisjoner += 1
```

```python
# Vi kan skrive en While-løkke som også itererer over elementene i en liste.

liste = ("eple", "banan", "tomat")

# Vi vil at listen skal kjøre like mange ganger som det er elementer i listen.
max = len(liste)

# Som vanlig bruker vi variabelen i til å telle hvor vi er i listen, vi starter på index 0.
i = 0

while i < max:
    element = liste[i]
    print(element)
    i += 1
```

### Oppgave 1.3.1.5
Pyramidetall n er summen av alle kvadrattall fra og med 1 til og med $ n^2 $. Pyramidetall 4 er  visualisert under.



(Bilde: Wikipedia)

Skriv et program som spør brukeren om hvilket tall hen ønsker, og bruker én løkke til å beregne tallet. Skriv ut tallet

```python
n = int(input("n: "))
sum = 0

i = 1
while i <= n:
    sum += i ** 2
    print("lag: " + str(i), "\tlag størrelse: " + str(i**2) + "\tsum: " + str(sum))
    i += 1

print("\nPyramide tall n =", sum)
```

### Oppgave 1.3.1.6

```python
# a)

i = 0
while i <= 5:
    print(i)
    i += 0.5
```

```python
# b)

i = 0
while i <= 5:
    print(i, "\t", round(i))
    i += 0.5
```

Jeg observerer at python bruker "Banker's rounding", som betyr at verdier under x.5 blir rundet ned og verdier over x.5 blir rundet opp, samt verdier som er lik x.5 blir rundet av til nermeste partall.

### Oppgave 1.3.2.5
Vi setter 10000 kr inn på en sparekonto med 10% rente årlig rente.

    a) Hvor mye penger er på kontoen etter 5 år?
    b) Hvor mange år må vi vente før det er over 20000 kr på kontoen?

```python
# a)

saldo = 10000
rente = 0.1

i = 0
while i < 5:
    saldo *= 1 + rente
    print(saldo)
    i += 1

print(round(saldo), "kr i kontoen etter 5 år")
```

```python
# b)

saldo = 10_000
rente = 0.1

i = 0
while saldo <= 20_000:
    saldo *= 1 + rente
    print(round(saldo, 2))
    i += 1

print("saldoen er over 20k etter", i, "år")
```

### Oppgave 1.3.2.6
På en øy har det brutt ut en sykdom i rottepopulasjonen. Det er per i dag 20000 rotter på øya, men populasjonen minker med 10% hvert år.

    a) Hvor mange rotter er det på øye om 3 år?
    b) Hvor mange år tar det før rottepopulasjonen er halvert?

```python
# a)

rotter = 20_000
endring = 0.9

i = 0
while i < 3:
    rotter *= endring
    i += 1

print("Populasjon etter 3 år", rotter)
```

```python
# b)

rotter_start = 20_000
rotter = rotter_start
endring = 0.9

i = 0
while rotter > rotter_start / 2:
    rotter *= endring
    i += 1

print("Populasjon halverte etter", i, "år.")
```

Skriv et program som bruker nøstede løkker til å vise følgende mønster:

\*\*\*\*\*\*\*\*\*\*<br>

\*\*\*\*\*\*\*\*<br>
\*\*\*\*\*\*\*<br>
\*\*\*\*\*\*<br>

```python
repetisjoner = 10

i = repetisjoner
while  i > 0:
    j = 0
    stjerner = ""
    while j < i:
        stjerner += "*"
        j += 1
    print(stjerner)
    i -= 1

```

### Oppgave 1.3.3.2
Skriv kode for å skrive ut følgende skjermbilde.

#

-##

--###

---####

----#####

-----######

------#######

-------########

```python
repetisjoner = 7

# Hovedløkke
i = 1
while i <= repetisjoner:
    j = 0
    streker = ""
    # Sub-løkke 1
    while j < i:
        streker += "-"
        j += 1
    
    k = 0
    tags = "#"
    # Sub-løkke 2
    while k < i:
        tags += "#"
        k += 1
    i += 1
    print(streker + tags)
```

### Oppgave 1.3.3.3
Skriv kode for å skrive ut følgende skjermbilde, ved hjelp av én ytre løkke og to indre løkker.
Ikke bruk if…else.

#

-##

--###

---####

#

```python
repetisjoner = 6

i = 1
while i < repetisjoner:
    l = -1 * abs(repetisjoner / 2 - i) + repetisjoner / 2
    
    j = 0
    streker = ""
    while j < l:
        streker += "-"
        j += 1
    
    k = 0
    tags = "#"
    while k < l:
        tags += "#"
        k += 1
    
    print(streker + tags)
    i += 1
```

### Oppgave 1.3.3.4
Skriv et program som viser følgende mønster:

33345

44445

55555

```python
repetisjoner = 5

i = 1
while i <= repetisjoner:
    j = 1
    tall = ""
    while j <= repetisjoner:
        if j > i:
            tall += str(j)
        else:
            tall += str(i)
        j += 1
    print(tall)
    i += 1
```
