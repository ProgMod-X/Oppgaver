### Deloppgave B
I programmet på arket brukte jeg dt = 0.001, men det går ikke siden linkingen jeg brukte er basert på at dt = 1. Dermed er den feil. Under er hvordan den burde ha vært. Vannmengden i reservoaret vil være 50 millioner kubikkmeter mellom 4 og 5 måned. Hvis vi legger på en round funksjon til vilkåret for while løkken vil den si at resorvoaret vil være på 50 millioner kubikkmeter på måned 4.

**Oppdatering**
Skulle ha ganget endringen med dt siden differensial likninker er kontinuerlige

```python
import matplotlib.pyplot as plt
import math

t = 0
dt = 0.0001
reservoar = 0

while round(reservoar) <= 50:
    # print(t, reservoar)

    reservoar += (-reservoar * 1/6 + 10 + 9 * math.sin(math.pi / 6 * t)) * dt
    t += dt
print(t, reservoar)
```

### Deloppgave E

```python
import matplotlib.pyplot as plt
import math

t = 0
dt = 0.0001
reservoar = 0
tmaks = 4 * 12
fastest_increase = (0, 0)

x_list =  []
y_list = []

while t <= tmaks:
    x_list.append(t)
    y_list.append(reservoar)

    tempydy = -reservoar * 1/6 + 10 + 9 * math.sin(math.pi / 6 * t)

    if tempydy > fastest_increase[1]:
        fastest_increase = (t, tempydy)

    dy = tempydy * dt
    
    
    reservoar += dy
    t += dt

print(f"Vannmengden økte raskest i måned {fastest_increase[0]} med {fastest_increase[1]} millioner kubikkmeter")

plt.plot(x_list, y_list)
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

t = 0
dt = 0.0001
reservoar = 0
tmaks = 4 * 12

x_list =  []
y_list = []
nabo_list = []

while t <= tmaks:
    x_list.append(t)
    y_list.append(reservoar)

    if reservoar > 20:
        nabo_list.append(reservoar - 20)
    else:
        nabo_list.append(0)

    dy = (-reservoar * 1/6 + 10 + 9 * math.sin(math.pi / 6 * t)) * dt
    reservoar += dy
    t += dt

sum = 0
for num in nabo_list:
    sum += num
average_amount_of_water_available = sum / len(nabo_list)

print(f"Nabo kommunen kan gjennomsnittlig ta {round(average_amount_of_water_available, 2)} millioner kubikkmeter vann per måned")

plt.plot(x_list, y_list)
plt.plot(x_list, nabo_list, "r")
plt.show()
```
