Opprett først filen temperatur.csv med dataene gitt under og importer den i koden din.

    a) Lag tre lister som representerer kolonnene i filen. Husk å fjerne unødvendige tegn og gjøre om datatyper.
    
    b) Plott et linjediagram uten punkter basert på tid og temperatur. La linjen være sort og stiplete (“--”).
    
    c) Plott punkter basert på temperaturen. Temperaturer under 0 grader skal ha blå prikker, temperaturer fra 0 og oppaver skal ha røde prikker.
    
    c) Beregn hvor stor del av døgnet det er meldt skyet vær og hvor stor del av døgnet det er meldt delvis skyet.

```python
# a)
file = open("temperatur.csv", "r")

file.readline()

times = []
weather = []
temps = []

for line in file:
    l = line.strip("\n").split(",")

    times.append(l[0])
    weather.append(l[1])
    temps.append(l[2])


file.close()
```

```python
# b & c)
import matplotlib.pyplot as plt

positive_temps = []
positive_times = []
negative_temps = []
negative_times = []

for i in range(len(temps)):
    if float(temps[i]) <= 0: # Skal egentlig være <, men vann fryser ved 0 så da bør den være blå
        negative_temps.append(temps[i])
        negative_times.append(times[i])
    else:
        positive_temps.append(temps[i])
        positive_times.append(times[i])


plt.plot(times, temps, "k--")
plt.plot(positive_times, positive_temps, "ro")
plt.plot(negative_times, negative_temps, "bo")
plt.show()
```

```python
partly_cloudy_count = 0
cloudy_count = 0
measurement_count = 0

for weather_instance in weather:
    if weather_instance == '"delvis skyet"':
        partly_cloudy_count += 1
    elif weather_instance == '"skyet"':
        cloudy_count += 1
    measurement_count += 1

partly_cloudy_percent = round((partly_cloudy_count / measurement_count) * 100)
cloudy_percent = round((cloudy_count / measurement_count) * 100)

print("partly cloudy:", partly_cloudy_percent, "%\ncloudy:", cloudy_percent, "%")
```
