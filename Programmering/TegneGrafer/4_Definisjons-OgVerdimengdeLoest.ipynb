# Definisjons og verdimengde
## Eksempel
> Tegn grafen til  $f\left(x \right) = \frac{1}{3} \; x^{3} - x^{2} + 2$ for definisjonsmengden $-2 \leq x \leq 3$. Bestem verdimengden til $f$.

Importerer grafbiblioteket, og lager to tomme lister.

```python
import matplotlib.pyplot as plt

x_liste = []
y_liste = []
```

Siden definisjonsmengden er $-2 \leq x \leq 3$, starter løkken på -2 og kjører til og med 3. 

```python
x = -2
while x <= 3:
    y = 1/3 * x**3 - x**2 + 2
    
    x_liste.append(x)
    y_liste.append(y)
    
    x += 0.1
```

Tegner grafen til $f$.

```python
plt.plot(x_liste, y_liste, "-m")
plt.show()
```

Finner minste og største _y_-verdi.

```python
minste_y  = min(y_liste)
print(minste_y)
```

```python
største_y = max(y_liste)
print(største_y)
```

## Oppgave 1
Funksjonen $f$ er gitt ved $f(x) = (x-1)(x-4)$ og har definisjonsmengden $D_f = [-1, 5]$.
- Tegn grafen til $f$.
- Bestem verdimengden til $f$, $V_f$.

```python
x_liste = []
y_liste = []

x = -1
while x <= 5:
    y = (x-1)*(x-4)
    
    x_liste.append(x)
    y_liste.append(y)
    
    x+= 0.1
    
plt.plot(x_liste, y_liste, "y")
plt.show()

y_min = min(y_liste)
y_max = max(y_liste)
print("Definisjonsmengde: [" + str(y_min) + ", " + str(y_max) +"]")
```

## Oppgave 2
Funksjonene  $f$ og $g$ er gitt ved $f(x)= x^2 - 4$ og $g(x) = 4 - x^2$.

Tegn grafene til funksjonene i samme koordinatsystem. Tilpass definisjonsmengden slik at skjæringspunktene mellom grafene vises.

```python
x_min     = -5
x_max     =  5
x_delta   = 0.1  # avstand mellom to x-verdier

x_liste   = []
y_liste_f = []
y_liste_g = []

x = x_min
while x <= x_max:
    f = x**2 - 4
    g = 4 - x**2
    
    x_liste.append(x)
    y_liste_f.append(f)
    y_liste_g.append(g)
    
    x += x_delta
    
plt.plot(x_liste, y_liste_f, 'm')
plt.plot(x_liste, y_liste_g, 'c')
plt.show()
```

## Oppgave 3
Funksjonen $f$ er gitt ved $f(x)=6 x - 12$, og verdimengden $V_f = [-12, 18]$.
- Bruk verdimengden i stedet for definisjonsmengden til å tegne grafen.
- Hvorfor går det her, men ikke for alle funksjoner?

Skriver om funksjonsuttrykket.
$$ y = 6 x - 12 \Leftrightarrow 6 x = y + 12 \Leftrightarrow x = \frac{y}{6} + 2 $$

```python
x_liste = []
y_liste = []

y = -12
while y<=18:
    x = y/6 + 2
    
    x_liste.append(x)
    y_liste.append(y)
       
    y += 0.1
    
plt.plot(x_liste, y_liste, 'k')
plt.show()
```
