## Oppgave 1
Tegn 100 tilfeldige punkter, der b√•de _x_- og _y_-verdiene ligger i intervallet $[0,100>$.

<details>
    <summary style="color:green;">üõà Klikk her for et hint</summary>
    Vi trekker et tilfeldig heltall med funksjonen <code>randint(start,slutt)</code>.
    For √• bruke denne funksjonen m√• du importere biblioteket <code>random</code>.
</details>

```python
import matplotlib.pyplot as plt
import random
```

Siden intervallene for _x_ og _y_ if√∏lge oppgaven inneholder 0 men ikke 100, m√• vi v√¶re obs p√• dette n√•r vi trekker tall.

I de to f√∏rste alternativene under trekkes tall i ‚Ñï‚ÇÄ.

### Alternativ 1

```python
x_liste = []
y_liste = []

punkt = 0
while punkt < 100:
    x_liste.append( random.randint(0,99) )
    y_liste.append( random.randint(0,99) )
  
    punkt += 1

plt.plot(x_liste, y_liste, "o")    
plt.show()
```

### Alternativ 2

```python
punkt = 0
while punkt < 100:
    x_liste = [ random.randint(0,99) ]
    y_liste = [ random.randint(0,99) ]
    
    plt.plot(x_liste, y_liste, "o")
    punkt += 1
    
plt.show()
```

### Alternativ 3
Her trekker vi f√∏rst et tilfeldig tall i intervallet $[0,1>$, og multipliserer med 100. 

Dette gj√∏r at vi f√•r koordinater i ‚Ñù fra og med 0 til men ikke med 100.

ü§î Hvilke andre endringer er gjort?

```python
punkt = 0
while punkt < 100:
    plt.plot( [ random.random() * 100 ], [ random.random() * 100 ], "s")
    punkt += 1
    
plt.show()
```

## Oppgave 2
Tegn 500 tilfeldige linjer. Hver linje skal trekkes mellom to punkter, der _x_-verdien skal ligge i intervallet $[0, 320>$ og _y_-verdien skal ligge i intervallet $[0,200>$.


<details>
    <summary style="color:green;">üõà Klikk her for et hint</summary>
    Her m√• du tegne en og en linje inne i en l√∏kke.
</details>

### Alternativ 1

```python
linje = 0
while linje < 500:
    x_liste = [ random.randint(0,319), random.randint(0,319) ]
    y_liste = [ random.randint(0,199), random.randint(0,199) ]
    
    plt.plot(x_liste, y_liste, "-")
    linje += 1
    
plt.show()
```

### Alternativ 2

ü§î Hvilke andre endringer er gjort?

```python
linje = 0
while linje < 500:
    x_liste = [ random.random() * 320, random.random() * 320 ]
    y_liste = [ random.random() * 200, random.random() * 200 ]
    
    plt.plot(x_liste, y_liste, ":")
    linje += 1
    
plt.show()
```
