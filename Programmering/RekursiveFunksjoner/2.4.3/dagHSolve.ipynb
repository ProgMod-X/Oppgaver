I en geometrisk tallfølge multipliserer vi det forrige leddet med en konstant for å få det neste.

 

I tallfølgen over er denne konstanten ½.


    a) Skriv en funksjon a(n) som bruker rekursjon til å returnere den n’te leddet i følgen.

Summer vi leddene i en tallfølge kalles det en tallrekke.

    b) Skriv et program som beregner summen av de n første leddene.
    c) Hva skjer når n blir veldig stor?
    Kan du tegne en figur som viser at det må være slik?

## a) og b)

```python
def a(n):
    if n == 0 or n == 1:
        return 1/2
    else:
        return a(n - 1) * 0.5

n = int(input("n: "))
sum = 0
for i in range(1, n + 1):
    sum += a(i)

print(sum)
```
