## a)

```python
import scipy.optimize as so
import matplotlib.pyplot as plt

data_aar = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007]
data_y = [383, 607, 832, 1081, 1193, 1352, 1667, 1905]
data_x = []

for val in data_aar:
    data_x.append(val - 2000)

def mal_f(x, a, b):
    return a * x + b

constants = so.curve_fit(mal_f, data_x, data_y)[0]
a = constants[0]
b = constants[1]

def f(x):
    return a * x + b

x_list = []
y_list = []

x_min = min(data_x)
x_max = max(data_x)

x = x_min
dx = 0.001

while x < x_max:
    x_list.append(x)
    y_list.append(f(x))
    
    x += dx

plt.plot(x_list, y_list, label=f"f(x) = {round(a, 2)} * x + {round(b, 2)}")
plt.xlabel("Antall år etter 2000")
plt.ylabel("Antall registrerte EL-biler i Norge")
plt.legend()
plt.show()
```

## b)

```python
print(f"2012: {f(2012 - 2000)}\t2014: {f(2014 - 2000)}")
```

I følge modellen var det 2911 EL-biler i 2012 og 3330 i 2014, men det var egentlig 9580 og 41051. Dette vil si modellen var ikke passer bra, og vi bør heller vurdere en eksponensiell/logistisk funksjon.
