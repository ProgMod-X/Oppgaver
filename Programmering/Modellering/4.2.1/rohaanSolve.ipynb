Formelen

$T = 2*\pi * \sqrt{\frac{L}{g}}$

kan brukes for å regne ut svingetiden til en pendel, når vi ser bort fra friksjon og luftmotstand. L  er snorlengden gitt i meter, og g er tyngdens akselerasjon. På jorden er g = 9.81 m/s2.

b) Gjør beregninger og sammenlikn uttrykket du fant for S(x) i oppgave a) med formelen for 

## a)

```python
import matplotlib.pyplot as plt
import scipy.optimize as so

data_x = [0.1, 0.3, 0.5, 0.8, 1.0, 1.3, 1.6, 2.0]
data_y = [0.69, 1.17, 1.44, 1.82, 2.08, 2.27, 2.53, 2.80]

def mal_f(x, a, b):
    return a * x**b

constants = so.curve_fit(mal_f, data_x, data_y)[0]
a = constants[0]
b = constants[1]

def f(x):
    return a * x**b

x_maks = max(data_x)
x = min(data_x)
dx = 0.001

x_list = []
y_list = []

while x < x_maks:
    x_list.append(x)
    y_list.append(f(x))

    x += dx

plt.plot(x_list, y_list, label=f"f(x) = {round(a, 2)} * x^{round(b, 2)}")
plt.xlabel("Pendel meter")
plt.ylabel("Svingetid")
plt.legend()
plt.show()
```

## b)

```python
import numpy as np

def T(x):
    return 2 * np.pi * np.sqrt(x / 9.81)

x_maks = max(data_x)
x = min(data_x)

x_list = []
y_list = []

while x < x_maks:
    x_list.append(x)
    y_list.append(T(x))

    x += dx

plt.plot(x_list, y_list, label=f"T(x) = 2 * pi * sqrt(L/g)")
plt.xlabel("Pendel meter")
plt.ylabel("Svingetid")
plt.legend()
plt.show()
```
