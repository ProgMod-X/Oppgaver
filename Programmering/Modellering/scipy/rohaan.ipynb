```python
import scipy.optimize as so
import matplotlib.pyplot as plt

data_aar = [1952, 1982, 1992, 2002, 2012, 2022]

data_x = []
for aar in data_aar:
    data_x.append(aar - 1950)

data_y = [65_956, 25_289, 19_780, 13_841, 9_825, 9_591]

def mal_f(x, a, b):
    return a / x + b

constants = so.curve_fit(mal_f, data_x, data_y)[0]
a = constants[0]
b = constants[1]

print(f"f(x) = {a}/x + {b}")

def f(x):
    return a / x + b

x_list = []
y_list = []

x_min = min(data_x)
x_maks = max(data_x)

dx = (x_maks - x_min) / 100
x = x_min

while x <= x_maks:
    x_list.append(x)
    y_list.append(f(x))
    x += dx

plt.plot(x_list, y_list, "-b")
plt.plot(data_x, data_y, "o")
plt.show()
```
