Bestem en lineær modell som viser denne sammenhengen.

$f(x) = a * x + b$

```python
import scipy.optimize as so
import matplotlib.pyplot as plt

data_x = [0.6, 2.5, 5.4, 7.8, 9.6]
data_y = [250, 480, 660, 920, 1140]

def mal_f(x, a, b):
    return a * x + b

constants = so.curve_fit(mal_f, data_x, data_y)[0]
a = constants[0]
b = constants[1]

print(f"f(x) = {a} * x + {b}")

def f(x):
    return a *x + b

x_list = []
y_list = []

x_min = min(data_x)
x_maks = max(data_x)

dx = (x_maks - x_min) / 100
x = x_min

while x <= x_maks:
    x_list.append(x)
    y_list.append(f(x))
    x += dx

plt.plot(x_list, y_list, "-b")
plt.plot(data_x, data_y, "o")
plt.show()
```
