#### a) 
Plott punktene i et koordinatsystem.
Punktene skal være blå.

```python
import matplotlib.pyplot as plt

x_list = [1, 13, 11, 4, 5, 8]
y_list = [1, 4, 9, 3, 6, 9]

plt.plot(x_list, y_list, "ob")
plt.show()
```

#### b)
Beregn og plott balansepunktet  i rødt.

```python
xsum = 0
ysum = 0
for num in x_list:
    xsum += num

for num in y_list:
    ysum += num

average_x = xsum / len(x_list)
average_y = ysum / len(y_list)

plt.plot(x_list, y_list, "ob")
plt.plot(average_x, average_y, "or")
plt.show()
```

#### c)
Bruk det punktet lengst til venstre og det punktet lengst til høyre til å beregne et midlertidig stigningstall for datasettet.

$a = \frac{(y - y_1)}{(x -x_1)}$

```python
x1_value = min(x_list)
x_value = max(x_list)

y1_value = y_list[x_list.index(x1_value)]
y_value = y_list[x_list.index(x_value)]

temp_growth_num = (y_value - y1_value) / (x_value - x1_value)
print(temp_growth_num)
```

#### d)
Bruk ettpunktsformelen sammen med balansepunktet og stigningstallet til å tegne en rett linje.

$y - y_1 = a(x -x_1)$

$y = a(x -x_1) + y_1$

Hvor punktet har verdiene $(x_1, y_1)$
I dette tilfellet ville det vært $(average_x, average_y)$

```python
def f(x):
    return temp_growth_num * (x - average_x) + average_y

new_x_list = []
new_y_list = []

dx = 0.1
x = 0
while x < max(x_list):
    new_x_list.append(x)
    new_y_list.append(f(x))

    x += dx

plt.plot(x_list, y_list, "ob")
plt.plot(average_x, average_y, "or")
plt.plot(new_x_list, new_y_list, "-k")
plt.show()
```
