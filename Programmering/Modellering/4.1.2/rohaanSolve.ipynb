## a)

```python
import scipy.optimize as so
import matplotlib.pyplot as plt

data_aar = [1970, 1980, 1990, 2000, 2010]
data_y = [4.6, 6.2, 7.6, 8.2, 9.5]
data_x = []
```

Trekk fra 1970 fra alle verdiene i **data_aar** siden x skal være antall år etter 1970

```python
for val in data_aar:
    data_x.append(val - 1970)
```

```python
def mal_f(x, a, b):
    return a * x + b

constants = so.curve_fit(mal_f, data_x, data_y)[0]
a = constants[0]
b = constants[1]

print(f"f(x) = {a} * x + {b}")

def f(x):
    return a * x + b

x_list = []
y_list = []

x_min = min(data_x)
x_max = max(data_x)

x = x_min
dx = 0.001

while x < x_max:
    x_list.append(x)
    y_list.append(f(x))

    x += dx

plt.plot(x_list, y_list, label=f"S(x) = {round(a, 2)} * x + {round(b, 2)}")
plt.xlabel("Antall år etter 1970")
plt.ylabel("Kilogram sjokolade per person")
plt.legend()
plt.show()
```
