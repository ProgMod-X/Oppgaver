# Oppgave 3.3.1.2
f(x) = cos(x) er den deriverte til en funksjon g(x), og g(0) = 0.

Du finner cosinusfunksjonen og pi i biblioteket math.

    a) Tegn grafen til f og g i samme koordinatsystem for intervallet x ∈ [0, 4 π].
    
    b) Hvilken sammenheng ser du mellom grafene til de to funksjonene?

```python
# a)
import matplotlib.pyplot as plt
from math import cos, pi

def f(x):
    return cos(x)

x_min = 0
x_max = 4 * pi
dx = 0.01

x_liste = []
y_liste = []
xd_liste = []
yd_liste = []

x = 0
y = 0 # F(-4) = =

while x < x_max:
    x_liste.append(x)
    y_liste.append(y)
    xd_liste.append(x)
    yd_liste.append(f(x))

    y += f(x) * dx
    x += dx

plt.plot(x_liste, y_liste, "r")
plt.plot(xd_liste, yd_liste, "b")
plt.show()
```
