# Oppgave 3.3.1.1
Under får du oppgitt ulike polynomfunksjoner f som er den deriverte til en funksjon F(x).

$ f(x)  = F'(x) $

Tegn grafen til funksjonen f (blå) sammen med den opprinnelige funksjonen F (rød) når

    a) f(x) = 0.5                           F(-4) = 0       x ∈ [-4, 4]

    b) f(x) = 2 x - 2                       F(0) = 0        x ∈ [0, 4]

    c) f(x) = 9 x^2 - 2 x - 2                F(-1) = -2      x ∈ [-1, 1]

    d) f(x) = 2 x^3 - 15 x^2 + 35 x - 25      F(1) = 0        x ∈ [1, 4]

Oppfølgingsspørsmål:

Hvilken sammenheng ser vi mellom grafene til funksjonene f og F.
Over har vi fått oppgitt et punkt på den opprinnelige grafen. Dersom vi bruker denne metoden uten dette punktet; hvilken informasjon går tapt og hva vet vi fremdeles sikkert om F?

```python
# a)
import matplotlib.pyplot as plt

def f(x):
    return 0.5

x_min = -4
x_max = 4
dx = 0.01

x_liste = []
y_liste = []
xd_liste = []
yd_liste = []

x = -4
y = 0 # F(-4) = 0

while x < x_max:
    x_liste.append(x)
    y_liste.append(y)
    xd_liste.append(x)
    yd_liste.append(f(x))

    y += f(x) * dx
    x += dx

plt.plot(x_liste, y_liste, "r")
plt.plot(xd_liste, yd_liste, "b")
plt.show()
```
