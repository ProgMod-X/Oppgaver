```python
import matplotlib.pyplot as plt

def df(x):         # df(x) = f'(x)
    y = 0.5
    return y

fra = -4
til =  4
y   =  0           # f(fra) = c

x_liste = []
y_liste = []

dx = 0.01

x = fra

while x <= til:
    x_liste.append(x)
    y_liste.append(y)

    y += df(x) * dx
    x += dx     
    
plt.plot(x_liste, y_liste)
plt.show()
```
