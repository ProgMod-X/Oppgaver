# Eulers metode
## Delmål
> Hvordan tegne grafen til en funksjon når vi kun har funksjonsuttrykket til den deriverte og en startverdi.

```python
import matplotlib.pyplot as plt
```

Først definerer vi den deriverte til en funksjon, et intervall $[fra, til]$. og en startverdi $c$. 

```python
def df(x):         # df(x) = f'(x)
    y = 2*x - 6
    return y

fra = -2
til =  8
c   =  0           # f(fra) = c
```

Bruker Eulers metode til å fylle og plotte punktlistene.

Metoden bruker det faktum at $f'(a)$ er stigningstallet til tangenten til grafen i $(a, f(a))$ til å estimere hvor grafen til $f$ skal ligge.



```python
x_liste = []
y_liste = []

dx = 0.01

y = c
x = fra
while x <= til:
    x_liste.append(x)
    y_liste.append(y)

    y += df(x) * dx
    x += dx     
    
plt.plot(x_liste, y_liste)
plt.show()
```
