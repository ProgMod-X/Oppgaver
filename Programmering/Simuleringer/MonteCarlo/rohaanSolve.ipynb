```python
import random
import math
import matplotlib.pyplot as plt

N = 10_000_000
x1, x2, y1, y2 = 0, 0, 0, 0
length = 1000
sum_of_distances = 0

x_values, y_values = [], []

for i in range(N):
    x1 = random.random() * length
    x2 = random.random() * length
    y1 = random.random() * length
    y2 = random.random() * length
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    sum_of_distances += distance

    x_values.append(i)
    y_values.append((sum_of_distances/(i + 1)))

plt.plot(x_values, y_values)
plt.show()

print(f"Estimert gjennomsnitlig distanse mellom to punkter i kvadrat med bredde og lengde {length}:\t{sum_of_distances/N}")
print(f"Gjennomsnitlig prosent av sidelengden {round(((sum_of_distances/N)/length) * 100, 1)}%")
```
