Hanne er kunstner og arbeider med rektangler i ulike strrelser. Lengdene til sidene i rektanglene er hele tall.

Hun vil dekke et rektangel med sm rde kvadrater. Kvadratene skal ligge tett inntil hverandre, og alle skal vre like store. Hanne vil lage en algoritme som hun kan bruke for  finne ut hvor store kvadratene maksimalt kan vre.

Hun har laget noen figurer som skal vise hvordan hun tenker.



    a) Lag en algoritme som beskriver mten Hanne vil lse problemet p.
    b) Bruk programmering til  lage et dataprogram med utgangspunkt i algoritmen.
    Nr Hanne forteller programmet hvor lange sidene i rektangelet er, skal programmet bestemme hvor store sidekantene i kvadratene maksimalt kan vre.

## a)
Vi starter med  fylle rektanglet med strst mulig kvadrater.

1. En av bredden eller lengden vil vre kortest, vi kan bruke den korteste siden til  lage et kvadrat som vil passe inn i rektanglet.
2. Hvis det er plass til flere av kvadratet legger man til flere.
3. Hvis er mer plass igjen men man fr ikke plass til kvadratet, fyller man den gjenvrende plassen ved  gjre steg 1 og 2 igjen p det nye rektanglet.
4. Hvis det ikke er mer plass igjen er kvadratet fullt, n m vi finne antall minste kvadrater som fr plass i rektanglet.
5. Antall av de minste kvadratene som ligger inntil de 2. minste kvadratene, ganget seg selv. Er hvor mange sm kvadrater man trenger for  erstatte de 2. minste.
6. gjenta steg 5 til kvadratet er fyllt av de sm kvadratene.

## b)

```python
from math import floor


def kvadraterIRektangel(lengde, bredde):
    kortestSide = min(lengde, bredde)
    lengsteSide = max(lengde, bredde)

    if lengsteSide % kortestSide == 0: # Hvis den lengste siden delt p den korteste er et heltall trenger vi ikke  fortsette
        areal = kortestSide ** 2
        return areal
    else:
        nyLengde = kortestSide
        nyBredde = lengsteSide - kortestSide

        areal = kvadraterIRektangel(nyBredde, nyLengde)

        return areal

l = float(input("Lengden av rektanglet: "))
b = float(input("Bredden av rektanglet: "))

a = kvadraterIRektangel(l, b)
print("Areal av de sm rektanglene: ", a)
```
