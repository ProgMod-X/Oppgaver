## a)

Brukeren setter inn lengdene av sidene i en trekant og programmet sjekker om trekanten er rettvinklet ved å bruke pytagoras læresetning.

## b)

```python
c = float(input("Skriv inn lengden til den lengste siden i trekanten"))
b = float(input("Skriv inn lengden til en annen side i trekanten"))
a = float(input("Skriv inn lengden til den siste siden i trekanten"))

if a**2 + b**2 == c**2:
    print("Trekanten er rettvinklet.")
else:
    print("Trekanten er ikke rettvinklet.")
```
