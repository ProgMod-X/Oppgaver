## a)

Hvis distansen mellom sentrum av sirkelen og punktet er:

    strre enn radiusen, vil punktet vre utenfor sirkelen
    mindre enn radiusen, vil punktet vre inni sirkelen
    lik radiusen, vil punktet vre p sirkelen

## b)

```python
a = float(input("x kordinatet til sentrum av sirkelen"))
b = float(input("y kordinatet til sentrum av sirkelen"))
r = float(input("radius til sirkelen"))

s = float(input("x kordinatet til punktet"))
t = float(input("y kordinatet til punktet"))

distanse =  ((s - a) ** 2 + (t - b) ** 2) ** 0.5

print("sirkel:", a, b, "radius:", r)
print("punkt:", s, t)

if distanse < r:
    print("Punktet ligger innenfor sirkelen.")
elif distanse == r:
    print("Punktet ligger p sirkelen")
else:
    print("Punkter ligger utenfor sirkelen")
```
